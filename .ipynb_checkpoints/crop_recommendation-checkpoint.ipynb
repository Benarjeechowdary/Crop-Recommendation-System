```python
def load_data(file_path,file_name):
    df=pd.read_csv(file_path+file_name)
    return df
```

```python
def scaling(data):
    x=data.drop(columns='label')
    a=x.columns.tolist()
    y=data['label']
    
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
    scaler=StandardScaler().fit(x_train)
    x_scaled=scaler.transform(x_train)
    x_test_scaled=scaler.transform(x_test)
    x_scaled_df=pd.DataFrame(x_scaled,columns=a)
    x_test_scaled_df=pd.DataFrame(x_scaled,columns=a)
    
    return x_scaled,x_test_scaled,y_train,y_test,scaler
```

```python
def model_train(x_train,y_train):
#     models={"Logistic Regression": LogisticRegression(max_iter=2000),
#         "Random Forest": RandomForestClassifier(),
#         "SVM": SVC(),
#         "KNN": KNeighborsClassifier()}
#     train_models={}
    
#     for name,model in models.items():
#         model.fit(x_train,y_train)
#         train_models[name]=model
    model=RandomForestClassifier(n_jobs=-1,random_state=42)
    train_model=model.fit(x_train,y_train)
    return train_model
```

```python
def get_best_model(model, X_test, y_test):
#     best_model = None
#     best_acc = 0
#     best_name = ""
    
#     for name, model in models.items():
#         preds = model.predict(X_test)
#         acc = accuracy_score(y_test, preds)
#         print(f"{name} Accuracy: {acc*100:.2f}%")
        
#         if acc > best_acc:
#             best_acc = acc
#             best_model = model
#             best_name = name
    y_pred=model.predict(X_test)
    model_evaluation=accuracy_score(y_pred,y_test)
    
    print("Model:",model,"Accuracy:",model_evaluation)
    return model,model_evaluation
```

```python
def predict_for_input(model,scaler,input_dict):
    df=pd.DataFrame([input_dict])
    scale=scaler.transform(df)
    prediction=model.predict(scale)[0]
    return prediction
    
```

```python
df=load_data(file_path,file_name)

X_train, X_test, y_train, y_test, scaler =scaling(df)

model=model_train(X_train,y_train)
best_model,acc = get_best_model(model, X_test, y_test)

```

```python
random_input = {
    "N": np.random.randint(0, 140),
    "P": np.random.randint(0, 140),
    "K": np.random.randint(0, 140),
    "temperature": round(np.random.uniform(10, 40), 2),
    "humidity": round(np.random.uniform(20, 100), 2),
    "ph": round(np.random.uniform(3, 10), 2),
    "rainfall": round(np.random.uniform(20, 300), 2)
}

print('random_input is:',random_input)

crop = predict_for_input(best_model, scaler, random_input)
print("\nBest Model Used:", best_model)
print("Predicted Crop:", crop)
```
